```python
class Board():
    
    def __init__(self):
        self.Queen = "Q"
        self.Blank = "-"
        self.board = [[self.Blank for _ in range(8)] for _ in range(8)]
        
    def __repr__(self):
        return "\n".join(" ".join(x) for x in self.board)
    
    def search(self):
        row = 0
        col = 0
        while row<8 and row>=0:
            if(self.is_valid_move(row,col)):
                self.set_queen_at(row,col)
                row += 1
                col = 0
            else:
                col += 1
                if(col>=8):
                    col = self.find_queen_on(row-1)+1
                    self.unset_queen_on(row-1)
                    row -= 1
        
        print("Found a solution!!")
        print(self)
        
    def is_valid_move(self, row, col):
        if(col>=8 or row>=8 or col<0 or row<0):
            return False
        if not self.is_valid_row(row,col):
            return False
        if not self.is_valid_col(row,col):
            return False
        if not self.is_valid_diag(row,col):
            return False
        return True
    
    def is_queen(self,row,col):
        return self.board[row][col] == self.Queen
    
    def is_on_board(self, row,col):
        return row>=0 and row<8 and col>=0 and col<8 
    
    def is_valid_row(self, row, col):
        for j in range(8):
            if self.is_queen(row,j):
                return False
        return True

    def is_valid_col(self, row, col):
        for i in range(8):
            if self.is_queen(i,col):
                return False
        return True

    def is_valid_diag(self, row, col):
        for i in range(8):
            if self.is_on_board(row-i,col-i) and self.is_queen(row-i,col-i) :
                return False
            if self.is_on_board(row+i,col-i) and self.is_queen(row+i,col-i) :
                return False
            if self.is_on_board(row-i,col+i) and self.is_queen(row-i,col+i) :
                return False
            if self.is_on_board(row+i,col+i) and self.is_queen(row+i,col+i) :
                return False
        return True

    
    
    def set_queen_at(self,row, col):
        self.board[row][col] = self.Queen
        
    def unset_queen_on(self,row):
        self.board[row] = ["-" for _ in range(8)]
        
    def find_queen_on(self,row):
        for col in range(8):
            if self.board[row][col] == "Q":
                return col
        raise ValueError("programmer error")
```

```python
board = Board()
board.search()
```
